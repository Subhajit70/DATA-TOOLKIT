```python
#3.Explain the following terms

# The difference in np.array, np.asarray and np.asanyarray

"""Use np.array() when you want to create a new ndarray from scratch.
Use np.asarray() when you want to convert input data to an ndarray but avoid unnecessary copying.
Use np.asanyarray() when you want to preserve the subclass type (if applicable) while converting to an ndarray"""

# the difference between deep copy and shallow copy.

"""Shallow copy shares references, while deep copy duplicates values.
Shallow copy is faster, but deep copy ensures independence between the original and the copy"""


import copy

# Original list
original_list = [1, 2, [3, 4]]

# Shallow copy
shallow_copy = copy.copy(original_list)

# Deep copy
deep_copy = copy.deepcopy(original_list)

# Modify the nested list in both copies
shallow_copy[2][0] = 99
deep_copy[2][0] = 88

print("Original list:", original_list)  # [1, 2, [99, 4]]
print("Shallow copy:", shallow_copy)    # [1, 2, [99, 4]]
print("Deep copy:", deep_copy)          # [1, 2, [88, 4]]

```

```python
#5.Create a NumPy array with random integers Petween 1 and 10 of shape (5,6 ))After creating the array

import numpy as np

# Generate a 5x6 array of random integers between 1 and 10
random_integers_array = np.random.randint(low=1, high=11, size=(5, 6))

print("Random integers array (shape 5x6):")
print(random_integers_array)




# Extract even integers from the array
even_integers = random_integers_array[random_integers_array % 2 == 0]

print("\nEven integers from the array:")
print(even_integers)



# Extract odd integers from the array
odd_integers = random_integers_array[random_integers_array % 2 != 0]

print("\nOdd integers from the array:")
print(odd_integers)
```

```python
#6.Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10

#a. Find the indices of the maximum values along each depth level (third axis
#b. Perform element-wise multiplication of between both arry


import numpy as np

# Generate a 3x3x3 array of random integers between 1 and 10
random_integers_3d = np.random.randint(low=1, high=11, size=(3, 3, 3))

print("Random integers 3D array (shape 3x3x3):")
print(random_integers_3d)


# Find indices of maximum values along the third axis
max_indices = np.argmax(random_integers_3d, axis=2)

print("\nIndices of maximum values along the third axis:")
print(max_indices)



# Create another 3x3x3 array (you can modify this array as needed)
another_array = np.random.randint(low=1, high=11, size=(3, 3, 3))

# Element-wise multiplication
result_array = np.multiply(random_integers_3d, another_array)

print("\nResult after element-wise multiplication:")
print(result_array)
```

```python
#7.Clean and transform the 'Phone' column in the sample dataset to remove non-numeric characters and convert it to a numeric data type. Also display the table attributes and data types of each column.



import pandas as pd

# Load the dataset (replace with the actual file path)
file_path = "People Data.csv"
df = pd.read_csv(file_path)

# Display the initial state of the DataFrame
print("Initial DataFrame:")
print(df.head())


# Clean the 'Phone' column by removing non-numeric characters
df['Phone'] = df['Phone'].str.replace('[^0-9]', '', regex=True)

# Convert the 'Phone' column to numeric (integer) data type
df['Phone'] = pd.to_numeric(df['Phone'], errors='coerce')

# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(df.head())


# Display DataFrame info (column names and data types)
print("\nDataFrame Info:")
print(df.info())
```

```python
#13.Create two NumPy arrays, x and y, each containing 100 random float values between 0 and 1. Perform the following tasks using Matplotlib and NumPy:

#a) Create a scatter plot using x and y, setting the color of the points to red and the marker style to 'o'.

#b) Add a horizontal line at y = 0.5 using a dashed line style and label it as 'y = 0.5'.

#c) Add a vertical line at x = 0.5 using a dotted line style and label it as 'x = 0.5'.

#d) Label the x-axis as 'X-axis' and the y-axis as 'Y-axis'.

#e) Set the title of the plot as 'Advanced Scatter Plot of Random Values'.

#f) Display a legend for the scatter plot, the horizontal line, and the vertical line


import numpy as np
import matplotlib.pyplot as plt

# Generate random float values between 0 and 1
x = np.random.rand(100)
y = np.random.rand(100)

# Create the scatter plot
plt.scatter(x, y, color='red', marker='o', label='Random Points')

# Add horizontal line at y = 0.5
plt.axhline(y=0.5, color='gray', linestyle='--', label='y = 0.5')

# Add vertical line at x = 0.5
plt.axvline(x=0.5, color='gray', linestyle=':', label='x = 0.5')

# Set axis labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Advanced Scatter Plot of Random Values')

# Display the legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
#14.Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:

#a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

#b) Label the x-axis as 'Date'.

#c) Set the title of the plot as 'Temperature and Humidity Over Time'

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic time-series data
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')
temperature = np.random.normal(loc=25, scale=5, size=100)
humidity = np.random.normal(loc=60, scale=10, size=100)

# Create the DataFrame
df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (Â°C)', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a secondary y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Set the title
plt.title('Temperature and Humidity Over Time')

# Display the legend
plt.legend(loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
#15.Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

#a) Plot a histogram of the data with 30 bins.

#b) Overlay a line plot representing the normal distribution probability density function (PDF).

#c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

#d) Set the title of the plot as 'Histogram with PDF Overlay

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate 1000 samples from a normal distribution
data = np.random.normal(size=1000)

# Create a histogram with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black', label='Histogram')

# Overlay the PDF (probability density function)
x = np.linspace(min(data), max(data), 100)
pdf = norm.pdf(x, loc=0, scale=1)
plt.plot(x, pdf, 'r', label='PDF')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
#16. Set the title of the plot as 'Histogram with PDF Overlay'.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate 1000 samples from a normal distribution
data = np.random.normal(size=1000)

# Create a histogram with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black', label='Histogram')

# Overlay the PDF (probability density function)
x = np.linspace(min(data), max(data), 100)
pdf = norm.pdf(x, loc=0, scale=1)
plt.plot(x, pdf, 'r', label='PDF')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
#17.Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'


import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data for demonstration
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)

# Determine quadrant based on x and y values
quadrant = np.where((x >= 0) & (y >= 0), 'I',
                    np.where((x < 0) & (y >= 0), 'II',
                             np.where((x < 0) & (y < 0), 'III', 'IV')))

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrant})

# Create the scatter plot
sns.scatterplot(x='x', y='y', hue='Quadrant', data=df, palette='Set1')

# Set labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Display the legend
plt.legend(title='Quadrant')

# Show the plot
plt.grid(True)
plt.show()


```

```python
#18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'


import numpy as np
from bokeh.plotting import figure, output_file, show

# Generate x values (time)
x = np.linspace(0, 2 * np.pi, 1000)

# Compute the sine wave function
y = np.sin(x)

# Create a Bokeh figure
p = figure(title='Sine Wave Function', x_axis_label='Time', y_axis_label='Amplitude')

# Add the sine wave line
p.line(x, y, line_width=2, legend_label='Sine Wave', line_color='blue')

# Add grid lines
p.grid.grid_line_alpha = 0.3

# Show the plot
output_file('sine_wave.html')
show(p)
```

```python
#19.Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'



import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.io import output_notebook

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 10, size=len(categories))

# Create a DataFrame
df = pd.DataFrame({'Category': categories, 'Value': values})

# Create a Bokeh figure
p = figure(x_range=categories, plot_height=400, title='Random Categorical Bar Chart')

# Add bars
source = ColumnDataSource(df)
p.vbar(x='Category', top='Value', source=source, width=0.5, color='skyblue', legend_label='Values')

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [('Category', '@Category'), ('Value', '@Value')]
p.add_tools(hover)

# Customize axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Show the plot
output_notebook()
show(p)


```

```python
#20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'


import plotly.express as px
import pandas as pd
import numpy as np

# Generate random data
x_values = np.linspace(0, 10, 100)
y_values = np.sin(x_values) + np.random.normal(0, 0.1, len(x_values))

# Create a DataFrame
df = pd.DataFrame({'X': x_values, 'Y': y_values})

# Create the line plot
fig = px.line(df, x='X', y='Y', title='Simple Line Plot')
fig.update_xaxes(title_text='X-axis')
fig.update_yaxes(title_text='Y-axis')

# Show the plot
fig.show()

```

```python
#21. Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'

import plotly.express as px
import pandas as pd
import numpy as np

# Generate random data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 10, size=len(categories))

# Create a DataFrame
df = pd.DataFrame({'Category': categories, 'Value': values})

# Create the pie chart
fig = px.pie(df, names='Category', values='Value', title='Interactive Pie Chart', labels={'Category': 'Categories'})

# Show the plot
fig.show()
```
